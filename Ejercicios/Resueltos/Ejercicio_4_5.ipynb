Suponga que los campos ``info`` de los nodos externos de un árbol contienen solo números y escriba una función que pueda invocarse como ``formula.evaluar()``, que al ser ejecutada entregue el valor numérico  de la fórmula representada por el árbol. Modifique a continuación la definición de las clases ``Arbol``, ``Nodoi`` y ``Nodoe`` para que incluyan la nueva función ``evaluar()``. Note que tiene que implementar la función ``evaluar()`` en las tres clases.

```python
class Nodoi:
    def __init__(self, izq, info, der):
        self.izq=izq
        self.info=info
        self.der=der

    def evaluar(self):
        valor_izquierda = self.izq.evaluar()  # obtenemos el valor de la izquierda
        valor_derecha = self.der.evaluar()    # obtenemos el valor de la derecha

        if self.info == "+":                               # operación si es suma
            return valor_izquierda + valor_derecha
        elif self.info == "-":                             # operación si es resta
            return valor_izquierda - valor_derecha
        elif self.info == "*":                             # operación si es multiplicacion
            return valor_izquierda * valor_derecha
        elif self.info == "/":                             # operación si es division
            return valor_izquierda / valor_derecha

class Nodoe:
    def __init__(self, info=""):
        self.info=info

    def evaluar(self):
        return self.info  # entrega el valor numerico

class Arbol:
    def __init__(self,raiz=Nodoe()):
        self.raiz=raiz

    def evaluar(self):
        return self.raiz.evaluar()     # opera la formula dentro del arbol
```

Pruébela a continuación:

```python
formula= Arbol(
            Nodoi(
                Nodoi(Nodoe(5),"+",Nodoe(2)),
                "*",
                Nodoi(
                    Nodoe(8),
                    "-",
                    Nodoi(Nodoe(9),"/",Nodoe(3))
                )
            )
        )
print(formula.evaluar())
```
