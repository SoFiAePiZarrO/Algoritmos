Se le llama "Camel Case" a la convención de escribir una frase sin espacios, pero marcando el inicio de cada palabra poniendo su primera letra en mayúscula. Por ejemplo, la frase
```
"  Algoritmos y    estructuras de   datos   "
```
transformada a Camel Case queda así:
```
"AlgoritmosYEstructurasDeDatos"
```

Escriba una función que transforme a Camel Case y pruébela:

````python
def CamelCase(s):
    """Retorna un string conteniendo la versión Camel Case del string s"""
    final = ""                                              # string vacío para formar la palabra final con cada palabra del input con la primera letra mayuscula
    for i in s.split():
      palabra = i.lower()                                   # ciclo for para cada palabra dentro de la frase input
      palabra_final = palabra[0].upper() + palabra[1:]      # hacemos mayuscula la primera letra de la palabra
      final = final + palabra_final                         # unimos la palabra nueva al final de la palabra que estamos formando
    return final
````

````python
print(CamelCase("    ALGORITMOS y    estructuras de   datos   "))
````

````python
print(CamelCase("    Algoritmos y  ESTRUCTURAS   "))
````
